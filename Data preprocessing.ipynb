```python
df1 = pd.read_csv("./data/daily_sales_data_0.csv")
df1
```

```python
df1 = df1[df1['product'] == 'pink morsel']
df1
```

```python
df1['price'] = df1['price'].replace('[\$,]','',regex=True).astype(float)
df1
```

```python
df1['sales'] = df1.apply(lambda row: row['price'] * row['quantity'], axis=1)
df1
```

```python
df1 = df1.drop('price',axis=1)
df1
```

```python
df1 = df1.drop('quantity', axis=1)
df1
```

```python
df1.to_csv('output1.csv', index=False)
```
